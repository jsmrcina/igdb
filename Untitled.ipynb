```python
import requests
import json
import time
import os.path

from twitch_token import twitch_token
from igdb_api import igdb_api
from IPython import display
from datetime import datetime, timedelta
```

```python
def queryIgdb(data, endpoint, delay = 0.3):
    if delay < 0.3 or delay > 100:
        print("Invalid delay, any delay less than 0.3 will hit a rate limit and more than 100s is unsupported")
        return None
        
    baseUri = "https://api.igdb.com/v4/"
    uri = baseUri + endpoint
    
    with open("twitch_secret.json", "r") as secret_file:
        with open("token.json", "r") as token_file:
            twitch_secret = json.loads(secret_file.read())
            twitch_token = json.loads(token_file.read())

            access_token = twitch_token["access_token"]
            headers = {
                "Client-ID": twitch_secret["client_id"],
                "Authorization": f"Bearer {access_token}",
                "Content-Type": "text/plain"
            }
    
    response = requests.post(uri, data=data, headers=headers)
    time.sleep(delay)
    return response

def queryGameById(id):
    data = f'fields *; where id = {id};'
    return queryIgdb(data, "games")

def queryGameByName(name):
    data = f'fields *; where name = "{name}";'
    return queryIgdb(data, "games")

def queryGenre(genre):
    data = f"fields *; where id = {genre};"
    return queryIgdb(data, "genres")

def queryPlatform(platform):
    data = f"fields *; where id = {platform};"
    return queryIgdb(data, "platforms")

def queryMultiplayerMode(mpMode):
    data = f"fields *; where id = {mpMode};"
    return queryIgdb(data, "multiplayer_modes")

def queryDevelopersByGameId(game):
    data = f"fields *; where developed = ({game});"
    return queryIgdb(data, "companies")

def queryPublishersByGameId(game):
    data = f"fields *; where published = ({game});"
    return queryIgdb(data, "companies")

def queryCoverByGameId(game):
    data = f"fields *; where game = ({game});"
    coverResponse = queryIgdb(data, "covers")
    highresImage = coverResponse.json()[0]["url"].replace("thumb", "1080p")[2:]
    uri = f"https://{highresImage}"
    return requests.get(uri).content

def printSection(sectionName, sectionList):
    print(f"{sectionName}: ", end="")
    print(', '.join(sectionList))
```

```python
token = twitch_token("twitch_secret.json")
token.queryTwitchAccessTokenFromFile()

gameName = input("Type game name: ")
print()

# Title
gameResponseFirst = queryGameByName(gameName).json()[0]
printSection("Title", [gameName])

# Release Date
release_date = [datetime.utcfromtimestamp(gameResponseFirst["first_release_date"]).strftime('%b %d, %Y')]
printSection("Release Date", release_date)

# Genres
genres = []
for genre in gameResponseFirst["genres"]:
    genres.append(queryGenre(genre).json()[0]["name"])
printSection("Genres", genres)

# Platforms
platforms = []
for platform in gameResponseFirst["platforms"]:
    platforms.append(queryPlatform(platform).json()[0]["name"])
printSection("Platforms", platforms)


# Multiplayer Modes
if "multiplayer_modes" in gameResponseFirst:
    for mpMode in gameResponseFirst["multiplayer_modes"]:
        mpModeResponse = queryMultiplayerMode(mpMode).json()[0]

        campaigncoop = mpModeResponse["campaigncoop"]
        lancoop = mpModeResponse["lancoop"]
        offlinecoop = mpModeResponse["offlinecoop"]
        onlinecoop = mpModeResponse["onlinecoop"]
        splitscreen = mpModeResponse["splitscreen"]

        multiplayer_modes = ["Single Player"]
        if campaigncoop or lancoop or offlinecoop or onlinecoop or splitscreen:
            multiplayer_modes.append("Co-op")
            multiplayer_modes.append("Multiplayer")

        printSection("Online", multiplayer_modes)
    
# Developers
developers = []
developerResponse = queryDevelopersByGameId(gameResponseFirst["id"])
for developer in developerResponse.json():
    developers.append(developer["name"])
printSection("Developers", developers)

# Publishers
publishers = []
publisherResponse = queryPublishersByGameId(gameResponseFirst["id"])
for publisher in publisherResponse.json():
    publishers.append(publisher["name"])
printSection("Publishers", publishers)

# Image
display.Image(queryCoverByGameId(gameResponseFirst["id"]))
```
